* Given a gate identity, able to determine gate rules associated with identity.* Gate rules are a generalized version of a gate identity.* An expression indicating that two circuits are equivalent, i.e. ABC = D.* Generated through four operations:  - LL - left circuit, leftmost gate  - LR - left circuit, rightmost gate  - RL - right circuit, leftmost gate  - RR - right circuit, rightmost gate  First letter indicates which circuit to focus on.  Second letter indicates which gate in the circuit to target on.  New rule or expression generated by multiplying both circuits by the inverse of the targeted gate.

```python
from sympy.physics.quantum.identitysearch import *from sympy.physics.quantum.gate import *
```

```python
x = X(0)y = Y(0)z = Z(0)h = H(0);cnot = CNOT(1,0)cgate_z = CGate((0,), Z(1))
```

```python
right_circuit = (h, cnot, cgate_z); right_circuit
```

Perform a LL operation:

```python
left, right = ll_op(left_circuit, right_circuit);
```
